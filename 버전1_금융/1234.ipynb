```python
import csv 
import numpy as np
import pandas as pd
```

```python
np_arr = np.loadtxt('data/test_data.CSV', delimiter=',', encoding='cp949', dtype=str)
np_arr
```

```python
# 데이터 프레임으로 바꾸기
# df = pd.DataFrame(np_arr)
df = pd.read_csv('data/test_data.CSV', encoding='cp949')
df
```

```python
# 결측지를 null로 채우기
df.fillna('NULL',inplace=True)
df
```

```python
# 계산을 위해 '나이'열을 숫자형으로 변환, 숫자로 변환안되면 nan으로 변환 
df['나이'] = pd.to_numeric(df['나이'], errors='coerce')
avg = df['나이'].mean()


# 평균나이와 각행의 나이 비교해서 차이계산 -> 새로운 df 생김
df['나이차이'] =  abs(df['나이']-avg)

# 비슷한 값 10개만 뽑아서 복사해서 새로운 dt 만들기 
df_avg10 = df.sort_values(by='나이차이').head(10).copy() 
#drop 은 보통 새객체반환/inplace=true하면 기존 dt를 수정함
df_avg10.drop(columns='나이차이',inplace=True)
df_avg10
```
